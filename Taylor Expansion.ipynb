```python
# Taylor formular.

# f(x) = sum of [infinity & n = 0] 1/n! * ( f**n(x0))(x - x0)**n.

def Taylor(x, x0, max):
    t = 0
    for n in range(max+1):
        t = t + np.exp(x0) * (x-x0)**n / np.math.factorial(n)
        return t
```

```python
Taylor(1, 2, 10)
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim([-5, 100])

x_list = np.linspace(-5,5,101)
plt.scatter(x_list, np.exp(x_list))
```

```python
# Lets plot Taylor.

max = 10
plt.plot(x_list, Taylor(x_list, 1, max), 'red')
```

```python
# Sine graph & function.

# sin(x) = sum of [infinity & n=0] (-1)**n / (2n + 1)! * x**(2n + 1)

def sin(x, max):
    t = 0 
    for n in range(max+1):
        t = t + (-1)**n * x**(2*n+1) / np.math.factorial(2*n+1)
    return t
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim([-2, 2])

x_list = np.linspace(-10,10,101)
plt.scatter(x_list, np.sin(x_list))
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim([-2, 2])

x_list = np.linspace(-10,10,101)
plt.scatter(x_list, np.sin(x_list))

plt.plot(x_list, sin(x_list, 3), 'red')
plt.plot(x_list, sin(x_list, 6), 'green')
plt.plot(x_list, sin(x_list, 9), 'yellow')
```
