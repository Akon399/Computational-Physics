```python
# Notes & Lecture by Akhona Njeje.
# Date : 18 August 2023.
# Topic : Higher Derivatives, 1st Principles formular & Taylor Expansion.



import numpy as np
import matplotlib.pyplot as plt

def derivative(f, x, h):
    return (f(x+h) - f(x)) / h
```

```python
# Calculate f(x).

def func(x):
    return 2*np.sin(x)**2 + x
```

```python
x0 = 10.5
h = 0.1

func(x0)
```

```python
derivative(func, x0, h)
```

```python
# Find the nth derivative.

def nDeriv(f, x, h, n):
    t = 0
    for k in range(n+1):
        t = t + (-1)**(k+n)*np.math.factorial(n) / (np.math.factorial(k)*np.math.factorial(n-k))*f(x+k*h)
        return t / h**n
```

```python
# n = nth devirative, n = 1 = 1st derivative.

nDeriv(func, x0, h, 1)
```

```python
# Taylor Expansion.

def taylor(f, x, x0, nmax, h):
    t = 0
    for n in range(nmax+1):
        t = t + nDeriv(f, x0, h, n) * (x - x0)**n / np.math.factorial(n)
    return t
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim([-2, 8])

x_list = np.linspace(-5,5,101)
plt.scatter(x_list, func(x_list))

nmax = 15
h = 0.05

plt.plot(x_list, taylor(func, x_list, 0, nmax, h), 'pink')
plt.plot(x_list, taylor(func, x_list, 2, nmax, h), 'blue')
plt.plot(x_list, taylor(func, x_list, -3, nmax, h), 'purple')
```

```python
# Interpolation.

def correctFunc(x):
    return 15 + 2.4*x - 0.5*x**2 - 0.35*x**3
# Polinomial to the power 3.
```

```python
npoints = 21
x_list = np.linspace(-5,5,npoints)
data = np.array([x_list, correctFunc(x_list)])
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(data[0], data[1])
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(data[0], data[1])
```

```python
# Linear & Cubic spines.

from scipy import interpolate as int

spline = int.interp1d(data[0], data[1], kind = 'linear')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(data[0], data[1])
plt.plot(data[0], spline(data[0])) # Spline is the function that connects the dots. 
```
